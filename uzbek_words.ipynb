```python
import requests
from nltk.corpus import wordnet
from bs4 import BeautifulSoup
url = "https://daryo.uz/"
user_agent = 'Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.114 Safari/537.36'
headers = { 'User-Agent' : user_agent }

response = requests.get(url, headers = headers)

soup = BeautifulSoup(response.text,'lxml')
links=[]
list_links = []
for link in soup.find_all('a'):
    links.append(link.get('href'))
    
for link in links:
    
    if link.startswith("/"):
        
        responses = requests.get("https://daryo.uz"+link, headers = headers)
        soups = BeautifulSoup(responses.text,'lxml')
        for i in soups.find_all('a'):
            #print(i.get('href'))
            list_links.append(i.get('href'))
#print(list_links)
```

```python
print(len(list_links))
```

```python
text = []  
my_uzbek_words = []
for link in list_links[:40]:
    if link.startswith("/"):
        #print("https://daryo.uz"+link)
        responses = requests.get("https://daryo.uz"+link, headers = headers)
        soups = BeautifulSoup(responses.text,'lxml')
        
        for script in soup(["script", "style"]):
            script.decompose()
        text.append(soups.get_text())    
        strips = list(soups.stripped_strings)
        f = []
        for i in text:
            #remove space
            i= i.strip()
            #print(i.strip())
            for j in i.split():
                #append latin characters to list 
                if j.isalpha() and j.isascii():
                    print(j)
                    f.append(j.lower())            
```
